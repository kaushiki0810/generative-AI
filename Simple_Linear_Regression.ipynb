```python
#find the independent and dependent feature
x = df.iloc[:,0] #independent feature
y = df.iloc[:,1] #dependent feature
x
```

```python
#plot the data
plt.scatter(df["cgpa"], df["package"])
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.show()
```

```python
#split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) #(to fix the shuffle process)
x_train
```

```python
#import the model and train
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train.values.reshape(-1, 1), y_train)
```

```python
m = model.coef_
c = model.intercept_
print("Slope:", m)
print("Intercept:", c) #or print(m, c)
```

```python
cgpa = 9.5
package = m * cgpa + c
print(" Package:", package)
```

```python
#predicting the model
y_pred = model.predict(x_test.values.reshape(-1, 1))
y_pred
```
